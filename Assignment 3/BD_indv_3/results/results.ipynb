```python
with open('formatted_results.csv', 'r', encoding="utf-8") as file:
    header = file.readline().strip().split(',')
    
    data = {'testParallelStreams':{}, 'testThreads':{}, 'testNaive':{}}
    
    for line in file:
        values = line.strip().split(',')
        key = values[0].split(':')[0].split('.')[1]
        if key not in data.keys():
            data[key] = {}
        
        measure = values[0].split(':')[1] if len(values[0].split(':')) > 1 else 'avgTime'

        if measure not in data[key]:
            data[key][measure] = {'size':[], 'score':[], 'error':[], 'units':[]}
        
        data[key][measure]['size'].append(values[1])
        data[key][measure]['score'].append(values[4])

        if(len(values)>6):
            data[key][measure]['error'].append(values[6])
            
        data[key][measure]['units'].append(values[-1])
```

```python
test_types = ['testParallelStreams', 'testThreads']
markers = ['o', 's', 'd']  # Different markers for each test type
colors = ['b', 'g', 'r']  # Different colors for each test type
labels = ['Parallel Streams', 'Threads']

# Initialize the plot
plt.figure(figsize=(10, 6))

# Loop over each test type and plot the data
for test_type, marker, color, label in zip(test_types, markers, colors, labels):
    sizes = [int(size) for size in data[test_type]['avgTime']['size']]
    scores = [float(score) for score in data[test_type]['avgTime']['score']]
    
    # Plot the data for this test type
    plt.plot(sizes, scores, marker=marker, linestyle='-', color=color, label=label)

# Add labels and title
plt.xlabel('Size')
plt.ylabel('Average time [ms/op]')
# plt.yscale('log')
# plt.xscale('log')

# Add grid, legend, and display the plot
plt.grid(True)
plt.legend()
plt.show()
```

```python
with open('threads.csv', 'r', encoding="utf-8") as file:
    header = file.readline().strip().split(',')
    
    data = {}
    
    for line in file:
        values = line.strip().split(',')
        key = values[1]

        if key not in data.keys():
            data[key] = {}
        
        size = int(values[2])
        if size not in data[key]:
            data[key][size] = {}
        
        measure = values[0].split(':')[1] if len(values[0].split(':')) > 1 else 'avgTime'

        if 'stack' in measure:
            continue
        
        if measure not in data[key][size]:
            data[key][size][measure] = {'score':[]}

        data[key][size][measure]['score'].append(float(values[5]))
```

```python
nThreads = data.keys()
colors = ['b', 'g', 'r', 'yellow', 'purple', 'pink', 'black', 'orange']  # Different colors for each test type

# Initialize the plot
plt.figure(figsize=(10, 6))

# Loop over each test type and plot the data
for nthreads, color in zip(nThreads, colors):
    sizes = list(data[nthreads].keys())
    scores = [data[nthreads][size]['avgTime']['score'][0] for size in sizes]
    print(sizes, scores)
    # Plot the data for this test type
    plt.plot(sizes, scores, marker='o', linestyle='-', color=color, label=nthreads)

# Add labels and title
plt.xlabel('Size')
plt.ylabel('Average time [ms/op]')
# plt.yscale('log')
# plt.xscale('log')

# Add grid, legend, and display the plot
plt.grid(True)
plt.legend()
plt.show()
```

```python
nThreads = data.keys()
colors = ['b', 'g', 'r', 'yellow', 'purple', 'pink', 'black', 'orange']  # Different colors for each test type

# Initialize the plot
plt.figure(figsize=(10, 6))

time_for_1 = []

# Loop over each test type and plot the data
for nthreads, color in zip(nThreads, colors):
    print(nthreads)
    sizes = list(data[nthreads].keys())
    scores = [data[nthreads][size]['avgTime']['score'][0] for size in sizes]
    if nthreads == '1':
        time_for_1 = scores
    print(time_for_1, scores)
    speedup = np.array(time_for_1) / np.array(scores)
    # Plot the data for this test type
    plt.plot(sizes, speedup, marker='o', linestyle='-', color=color, label=nthreads)

# Add labels and title
plt.xlabel('Size')
plt.ylabel('Speedup per thread')

# Add grid, legend, and display the plot
plt.grid(True)
plt.legend()
plt.show()
```
