```python
import matplotlib.pyplot as plt
```

```python
nThreads = [2, 4, 8, 16, 32]
avgTime = [48.142, 41.713, 38.066, 40.987, 55.054]

plt.plot(nThreads, avgTime)
plt.xlabel("number of threads")
plt.ylabel("average time [ms/op]")
plt.show()
```
